```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import operator
import statistics
import re
%matplotlib inline
```

```python
air_data = pd.read_excel('AirQualityUCI.xlsx')
```

```python
air_data.head()
label = ['CO' , 'PT08.S1(CO)' , 'NMHC(GT)' , 'C6H6(GT)' ,'PT08.S2(NMHC)' , 'NOx(GT)' , 'PT08.S3(NOx)' , 'NO2(GT)' , 'PT08.S4(NO2)' , 'PT08.S5(O3)' , 'T' , 'RH' , 'AH']
m=["mar 2004","apr 2004","may 2004","jun 2004","jul 2004","aug 2004","sep 2004","oct 2004","nov 2004","dec 2004","jan 2005","feb 2005","mar 2005","apr 2005"]
df=pd.DataFrame(air_data)
```

```python
air_data.dropna(axis=0, how='all')
```

```python
from pandas import Timestamp
months = [g for n, g in df.set_index('Date').groupby(pd.TimeGrouper('M'))]
```

```python

monn=[]
for i in range(0,len(months)):
    b=[]
    a=months[i].mean(axis=0)
    b.append(a)
    monn.append(b)

   
print(monn)
```

```python
s=[]
for i in range(0,len(monn)):
    num1 = []
    f1=[]
    result=[]
    monn[i]=str(monn[i])
    result = re.findall(r"[-+]?\d*\.\d+|\d+", monn[i])
    for i in result:
        if "." in i:
            num1.append(i)
    #print(num1)
    for i in num1:
        e=float(i)
        f1.append(e)
    #print(f1)
    d1 = dict(zip(label,f1))
    #print(d1)
    s1 = dict(sorted(d1.items(), key=operator.itemgetter(1), reverse=True))
    #print(s1)
    s.append(s1)
    
    #plt.bar(range(len(monn[i])),values1,tick_label=names1)
    #plt.show()
j=0
for i in s:
    names1 = list(i.keys())
    values1 = list(i.values())
    plt.title(m[j])
    j+=1
    
    plt.barh(range(len(i)),values1,tick_label=names1)
    plt.show()
```

```python
features = air_data
```

```python
features = features.drop('Date', axis=1)
features = features.drop('Time', axis=1)
features = features.drop('PT08.S2(NMHC)', axis=1)
```

```python
labels=air_data['PT08.S2(NMHC)'].values
```

```python
features=features.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3)
```

```python
print("X_trian shape --> {}".format(X_train.shape))
print("y_train shape --> {}".format(y_train.shape))
print("X_test shape --> {}".format(X_test.shape))
print("y_test shape --> {}".format(y_test.shape))
```

```python
### Linear  Regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print("Predicted values:", regressor.predict(X_test))
```

```python
print("Coefficient of determination R^2 <-- on train set: {}".format(regressor.score(X_train, y_train)))
```

```python
print("R^2 score for liner regression: ", regressor.score(X_test, y_test))
```

```python
#SVM
```

```python
from sklearn.model_selection import KFold
from sklearn.svm import SVR
```

```python
support_regressor = SVR(kernel='rbf', C=1000)
support_regressor.fit(X_train, y_train)
```

```python
print("Coefficient of determination R^2 <-- on train set: {}".format(support_regressor.score(X_train, y_train)))
```

```python
print("Coefficient of determination R^2 <-- on test set: {}".format(support_regressor.score(X_test, y_test)))
```

```python
#DECISION TREE
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
```

```python
print("Coefficient of determination R^2 <-- on train set: {}".format(dtr.score(X_train, y_train)))
```

```python
print("Coefficient of determination R^2 <-- on test set: {}".format(dtr.score(X_test, y_test)))
```

```python
#FEATURE SELECTION
```

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
etr = ExtraTreesRegressor(n_estimators=300)
etr.fit(X_train, y_train)
```

```python

ar=etr.feature_importances_
print(ar)
indecis = np.argsort(ar)[::-1]
print(indecis)
l=[]
for i in indecis:
    l.append(label[i])
print(l)
d2 = dict(zip(l,ar))
#print(d2)
s2 = dict(sorted(d2.items(), key=operator.itemgetter(1),reverse=True))
print(s2)
```

```python
names1 = list(s2.keys())
values1 = list(s2.values())
```

```python
plt.figure(num=None, figsize=(19, 10), dpi=80, facecolor='w')
plt.title("Feature importances")
plt.bar(range(len(s2)),values1,tick_label=indecis,color="r",align="center")
#plt.savefig('bar.png')
#plt.show()
```
